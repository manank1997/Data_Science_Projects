```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import solve_ivp
import matplotlib as mpl

#Optional, Remove all the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def Henon(x,y,a,b):
    return [a - x**2 + b*y, x]
```

```python
a = 1.2;
b = 0.4;
x = 0.;
y = 0.;

fir = plt.figure(figsize = (12,12))

for i in range(0,100):
    [x,y] = Henon(x,y,a,b)

for i in range(0,2000):
    [x,y] = Henon(x,y,a,b)
    plt.plot(x,y,'r.', ms = 2.5)

plt.xlim(-2.4,2.4);
plt.ylim(-2.45,2.45);
plt.xlabel('$x_n$', fontsize = 32);
plt.ylabel('$y_n$', fontsize = 32);
```

```python
#Henon's Map Basin Boundaries

xgrid = np.arange(-2.5,2.5,0.01)
ygrid = np.arange(-7.5,7.5,0.01)

a=1.28;b=0.4
#a=2.0;b=-0.3

Xbb = np.zeros((len(ygrid),len(xgrid)))

l = 0
for x in xgrid:
    m = 0;
    x0 = x
    for y in ygrid:
        r = 0.0;
        x = x0
        for i in range(10):
            [x,y] = Henon(x,y,a,b)
            r = np.sqrt(x**2 + y**2)
            if (r>2):
                Xbb[m,l] = i
        m += 1
    l += 1
```

```python
fig = plt.figure(figsize = (20,10))
plt.imshow(Xbb, origin = 'lower', cmap = 'magma_r', extent= [min(xgrid), max(xgrid), min(ygrid), max(ygrid)], aspect = 0.3)
plt.clim(0,10)
plt.colorbar()
```
