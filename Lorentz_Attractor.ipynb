```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import solve_ivp
import matplotlib as mpl

#Optional, Remove all the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def lorenz(Var,t,sig,b,r):
    x,y,z = Var
    x_dot = sig*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot
```

```python
sig = 10
b = 2.667
r = 28
```

```python

#Initial Conditions
Var0 = [-0.75, 0.25, 0.1]

#Time Steps
t = np.linspace(0,100,100000)

sol = solve_ivp(lambda t, Var: lorenz(Var, t, sig, b, r), [t[0],t[-1]], Var0, method = 'RK45', t_eval = t)
```

```python
# Plot the Lorenz Attractor

fig = plt.figure(figsize=(15, 10))
ax = fig.gca(projection='3d')

mpl.rcParams['axes.labelsize'] = 36
mpl.rcParams['text.usetex']= 'False'

mpl.rcParams['xtick.labelsize'] = 36
mpl.rcParams['ytick.labelsize'] = 36



ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))


ax.set_xlabel("X(t)")
ax.set_ylabel("Y(t)")
ax.set_zlabel("Z(t)")

ax.plot(sol.y[0,:],sol.y[1,:],sol.y[2,:],linewidth=2, color=[0.3,0.5,0.7])

ax.set_xticks([-20,20])
ax.set_yticks([-20,20])
ax.set_zticks([0,40])
ax.grid(False)

#ax.view_init(10, 250)

plt.show()
```
