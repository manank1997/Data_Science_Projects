```python
import skdim
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from sklearn import manifold, datasets
```

```python
sr_points , sr_color = datasets.make_swiss_roll(n_samples = 5500, random_state= 0)
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(projection= '3d')
fig.add_axes(ax)
ax.scatter(sr_points[:, 0], sr_points[:, 1], sr_points[:, 2], c=sr_color, s=10, alpha=1.0,cmap='inferno')
ax.set_title("Swiss Roll", fontsize=50)

ax.view_init(azim=66, elev=10)
```

```python
corr = skdim.id.CorrInt(k1 = 1, k2 = 2, DM=False)
dim_i = corr.fit(sr_points).dimension_
print('Intrinsic Dimenstion Using Correlation Dimension:', dim_i)
```

```python
dnn = skdim.id.KNN(k = 10, ps = None, M = 50, gamma = 2)

dim_i = dnn.fit(sr_points).dimension_
print('Intrinsic Dimension Using Nearest Neighbor Dimension:', dim_i)
```

```python
id_1 = skdim.id.MADA(DM = False)
dim_i = id_1.fit(sr_points).dimension_
print('Intrinsic Dimension Using Information Dimension:', dim_i)
```
