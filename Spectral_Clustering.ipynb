```python
#import packages we need 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
```

```python
# See scikit learn documentation for more details 

n_samples = 500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=0.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
# Anisotropicly distributed data
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)
```

```python
plt.figure(figsize=(20,5))


plt.subplot(1,4,1) 
plt.title('Noisy Circles', fontsize=20)
plt.scatter(noisy_circles[0][:,0],noisy_circles[0][:,1],marker='o',s=30,color=[0.5,0.5,0.5])
plt.xticks(());plt.yticks(());

plt.subplot(1,4,2) 
plt.title('Noisy Moons', fontsize=20)
plt.scatter(noisy_moons[0][:,0],noisy_moons[0][:,1],marker='o',s=30,color=[0.5,0.5,0.5])
plt.xticks(());plt.yticks(());

plt.subplot(1,4,3)
plt.title('Blobs', fontsize=20) 
plt.scatter(blobs[0][:,0],blobs[0][:,1],marker='o',s=30,color=[0.5,0.5,0.5])
plt.xticks(());plt.yticks(());

plt.subplot(1,4,4)
plt.title('Anisotropicly Distributed', fontsize=20) 
plt.scatter(aniso[0][:,0],aniso[0][:,1],marker='o',s=30,color=[0.5,0.5,0.5])
plt.xticks(());plt.yticks(());
```

```python
from sklearn.cluster import SpectralClustering

data_= noisy_circles[0]
clustering = SpectralClustering(n_clusters=2,affinity="nearest_neighbors",random_state=1).fit(data_)
label_noisy_circles=clustering.labels_



data_= noisy_moons[0]
clustering = SpectralClustering(n_clusters=2,affinity="nearest_neighbors",random_state=1).fit(data_)
label_noisy_moons=clustering.labels_

data_= blobs[0]
clustering = SpectralClustering(n_clusters=3,affinity="nearest_neighbors",random_state=1).fit(data_)
label_blobs=clustering.labels_

data_= aniso[0]
clustering = SpectralClustering(n_clusters=3,affinity="nearest_neighbors",random_state=1).fit(data_)
label_aniso=clustering.labels_





plt.figure(figsize=(20,10))

plt.title('Spectral Clustering')
cmap = plt.get_cmap('Dark2_r', 5)




plt.subplot(2,4,1) 
plt.title('Noisy Circles', fontsize=20)
plt.scatter(noisy_circles[0][:,0],noisy_circles[0][:,1],marker='o',s=30,c=label_noisy_circles,cmap=cmap)
plt.xticks(());plt.yticks(());

plt.subplot(2,4,2) 
plt.title('Noisy Moons', fontsize=20)
plt.scatter(noisy_moons[0][:,0],noisy_moons[0][:,1],marker='o',s=30,c=label_noisy_moons,cmap=cmap)
plt.xticks(());plt.yticks(());

plt.subplot(2,4,3)
plt.title('Blobs', fontsize=20) 
plt.scatter(blobs[0][:,0],blobs[0][:,1],marker='o',s=30,c=label_blobs,cmap=cmap)
plt.xticks(());plt.yticks(());

plt.subplot(2,4,4)
plt.title('Anisotropicly Distributed', fontsize=20) 
plt.scatter(aniso[0][:,0],aniso[0][:,1],marker='o',s=30,c=label_aniso,cmap=cmap)
plt.xticks(());plt.yticks(());

from sklearn.cluster import KMeans


data_= noisy_circles[0]
clustering = KMeans(n_clusters=2).fit(data_)
label_noisy_circles=clustering.labels_


data_= noisy_moons[0]
clustering = KMeans(n_clusters=2).fit(data_)
label_noisy_moons=clustering.labels_

data_= blobs[0]
clustering = KMeans(n_clusters=3).fit(data_)
label_blobs=clustering.labels_

data_= aniso[0]
clustering = KMeans(n_clusters=3).fit(data_)
label_aniso=clustering.labels_








plt.subplot(2,4,5) 
plt.scatter(noisy_circles[0][:,0],noisy_circles[0][:,1],marker='o',s=30,c=label_noisy_circles,cmap=cmap)
plt.xticks(());plt.yticks(());

plt.subplot(2,4,6) 
plt.scatter(noisy_moons[0][:,0],noisy_moons[0][:,1],marker='o',s=30,c=label_noisy_moons,cmap=cmap)
plt.xticks(());plt.yticks(());

plt.subplot(2,4,7)
plt.scatter(blobs[0][:,0],blobs[0][:,1],marker='o',s=30,c=label_blobs,cmap=cmap)
plt.xticks(());plt.yticks(());

plt.subplot(2,4,8)
plt.scatter(aniso[0][:,0],aniso[0][:,1],marker='o',s=30,c=label_aniso,cmap=cmap)
plt.xticks(());plt.yticks(());


print('Top panel is Spectral Clustering and bottom panel is Kmeans clustering')
```
